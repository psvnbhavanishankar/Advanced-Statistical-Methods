```
import random
import math
import statistics
import matplotlib.pyplot as plt


_lambda = 8.67
_num_events = 200000

_event_num = []
_inter_event_times = []
_event_times = []
_event_time = 0

print('EVENT_NUM,INTER_EVENT_T,EVENT_T')

```

```
for i in range(_num_events):
  
  n = random.random()

  _inter_event_time = -math.log(1.0 - n) / _lambda
  

  _event_time = _event_time + _inter_event_time
  if math.floor(_event_time) in _event_times:
    continue
  else:
    _event_times.append(math.floor(_event_time))
    _inter_event_times.append(_inter_event_time)
    _event_num.append(i)

  if len(_event_times) >= 59:
    print(len(_event_times))
    break

  print(str(i) +',' + str(_inter_event_time) + ',' + str(_event_time))
```
